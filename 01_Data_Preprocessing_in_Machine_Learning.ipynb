```python
import pandas as pd
data = {
    'Name' : ['Pravan','Kapil','Kanchan', 'Kratika','Ishan','Om'],
    'Age' : [25, None , 44 ,23 , None, 50 ],
    'Salary' : [5000,6000,7000,5000, None, None]
}
df  = pd.DataFrame(data)
print("Original Data Frame")
print(df)
```

```python
print(df.isnull().sum())
```

```python
#for dropping the values  - None values
df_drop = df.dropna()
print(df_drop)
```

```python
#for filling the avg values inside insead of drop none values
df['Age'].fillna(df['Age'].mean() , inplace  = True)
df['Salary'].fillna(df['Salary'].mean(), inplace = True )
print(df)
```

for finding the missing data in %
print(df.isnull().mean())*100

```python
print(df.isnull().mean()*100)
```

#Encoding Categorical Values
1> Label Encoding,
2> One - hot Encoding

```python
from sklearn.preprocessing import LabelEncoder # from sklearn library we want label encoder tool (which we are taking from preprocessing)
import pandas as pd
df = pd.read_excel("/content/sample_data.csv.xlsx", header=None)
df.columns = ['Name', 'Gender', 'City', 'Passed'] # Assign explicit column names

# Clean column names by stripping whitespace (though less critical after manual assignment)
df.columns = df.columns.str.strip()

print("DataFrame Columns after loading and stripping:", df.columns)

df_label = df.copy ()#duplicate data from original data , so original table don't change
le = LabelEncoder()#making obj
df_label['Gender_Encoder'] = le.fit_transform(df_label['Gender'])
df_label['Passed_Encoder'] = le.fit_transform(df_label['Passed'])
print('\n Label Encoded data')
print (df_label [['Name','Gender','Gender_Encoder','Passed','Passed_Encoder']])# for 2d data [[]] if we don' t use [[]] it ll like series data
#head for starting 5 data

df_encoded = pd.get_dummies(df, columns=['City']) #dummies -> tool in pandas -> text colum to small binary conversion
#and one column describe one category
print('\n One-Hot Encoded data (City)')
print(df_encoded)
```
